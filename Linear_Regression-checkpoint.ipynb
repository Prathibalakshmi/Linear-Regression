```python
import numpy as np
import matplotlib.pyplot as plt
```

## Plotting a line for a given (a,b)

```python
def get_line(a,b,x):
    y = a*x + b
    return y
```

```python
a1 = 3; b1 = 5
a2 = 5; b2 = 10
x = np.linspace(0,10,100)
y1 = get_line(a1,b1,x)
y2 = get_line(a2,b2,x)
```

```python
plt.figure(figsize=(9,6))
plt.plot(x,y1,label='a={},b={}'.format(a1,b1))
plt.plot(x,y2,label='a={},b={}'.format(a2,b2))
plt.title("Y = aX + b")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

## Line equation from 2 points

```python
p1 = [-15,-40]; p2 = [20,80]
```

```python
def get_line_from_2_points(p1,p2,x):
    x1, y1 = p1; x2, y2 = p2;
    m = (y2 - y1)/(x2 - x1)
    y = y1 + m*(x-x1)
    return y
```

```python
x = np.linspace(-25,25,100)
y = get_line_from_2_points(p1,p2,x)
```

```python
plt.figure(figsize=(9,6))
plt.plot(x,y)
plt.plot(p1[0],p1[1],'o',label="{}".format(p1))
plt.plot(p2[0],p2[1],'o',label="{}".format(p2))
plt.title("Y = aX + b")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

## Generate noisy dataset

```python
def get_random_vector_in_range(a,b,n):
    arr = a + (b-a)*np.random.random(n)
    return arr 
```

```python
def generate_noisy_dataset(a,b,x,sig):
    n = len(x)
    y = a*x + b
    rand_array = get_random_vector_in_range(-sig,sig,n)
    y_noisy = y + rand_array
    # y_noisy = np.array([y[i]+rand_array[i] for i in range(n)])
    return x, y, y_noisy
```

```python
x = np.linspace(-10,10,100)
a = -4; b = 5; 
sig = 2
x, y, y_noisy = generate_noisy_dataset(a,b,x,sig)
```

```python
plt.figure(figsize=(9,6))
plt.plot(x,y,color='red',label='original line')
plt.scatter(x,y_noisy,color='blue',s=10,label='noisy points')
plt.title("Y = {}X + {} + sig".format(a,b))
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

## Line fitting using partial derivatives

```python
X = np.matrix([x,y_noisy]).T   ## noisy dataset
X.shape
```

```python
mu_x = np.mean(X[:,0])
mu_y = np.mean(X[:,1])
mu_x2 = np.mean(np.multiply(X[:,0],X[:,0]))   # x**2 element-wise   
mu_xy = np.mean(np.multiply(X[:,0],X[:,1]))   # element-wise multiplication 
```

```python
m = [2,5,6]
n = [3,6,3]

np.multiply(m,n)
```

### Simultaneous Equation Solving

```python
a_approx = (mu_xy - mu_x * mu_y)/(mu_x2 - (mu_x)**2)
b_approx = mu_y - a*mu_x
```

```python
print("Acutal value of a={}; Approximated value of a={}".format(a,a_approx))
print("Acutal value of b={}; Approximated value of b={}".format(b,b_approx))
```

### Matrix Method

```python
A = np.matrix([[1, mu_x],[mu_x,mu_x2]])
A_inv = np.linalg.inv(A)
y = np.matrix([mu_y, mu_xy]).T
```

```python
params = A_inv.dot(y)
params
```

```python
params.shape
```

```python
b_approx = params[0,0]; a_approx = params[1,0]
print("Acutal value of a={}; Approximated value of a={}".format(a,a_approx))
print("Acutal value of b={}; Approximated value of b={}".format(b,b_approx))
```

### Example Exercise

```python
x = [-1.313, 8.585, 5.353, 10, -8.383]
y = [-1.062, 39.506, 27.089, 46.69, -28.019]

X = np.matrix ([x, y]).T
mu_x = np.mean(X[:,0])
mu_y = np.mean(X[:,1])
mu_x2 = np.mean(np.multiply(X[:,0], X[:,0])) # x**2 element-wise
mu_xy = np.mean(np.multiply(X[:,0], X[:,1])) # Elementwise multiplication
A = np.matrix([[1, mu_x], [mu_x, mu_x2]])
A_inv = np.linalg.inv(A)
y = np.matrix([mu_y, mu_xy]).T
params = A_inv.dot(y)
b_approx = params[0, 0]; a_approx = params[1,0]
print("Actual value of a={}; Approximated value of a = {}".format(a, a_approx))
print("Actual value of b={}; Approximated value of b = {}".format(b, b_approx))
```

## Mean Square Error(MSE)

```python
y_approx = a_approx * x + b_approx
```

```python
def MSE(y_approximated, y_noisy):
    e = y_noisy - y_approximated
    mse = np.mean(e*e)   # element-wise operation
    return mse
```

```python
MSE(y_approx, y_noisy)
```

```python
## Library function for Mean Squared Error
from sklearn.metrics import mean_squared_error
mean_squared_error(y_approx, y_noisy)
```

## R-Square Score

```python
def my_r2_score(y_approx, y_noisy):
    mu_y = np.mean(y_noisy)
    num =  np.sum((y_noisy - y_approx)**2)
    deno = np.sum((y_noisy - mu_y)**2)
    return (1 - num/deno)
```

```python
my_r2_score(y_approx, y_noisy)
```

```python
## Library function for r2_score
from sklearn.metrics import r2_score
r2_score(y_approx, y_noisy)
```
